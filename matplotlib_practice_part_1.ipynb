```python
#******************LINE CHART********************
# MatPLot and Object oriented Methods
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Method #1 MATLAB method
# Create the plot CHAIN THE plot() FUNCTION TO plt
plt.plot(x_axis,y_axis)

plt.show()
```

```python
# Method #2 OBJECT-ORIENTED Interface method
# Create the plot with ax.plt()
fig, ax = plt.subplots() 
ax.plot(x_axis, y_axis)  
    #explanation: 
        # plt.subplots() returns a TUPLE that contains a figure and axes object(s)
        # fig, ax = plt.subplots() unpacks the tuple into the variables fig and ax. 
        # The first variable, fig, references the figure.
        # The second variable, ax, references the axes of the figure and allows us to annotate the figure.
        # The ax.plot(x_axis, y_axis) plots the data from the x and y axes.
        
plt.show()
```

```python
# fig, ax = plt.subplots() line of code can also be written like this.
# Create the plot with ax.plt() 
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)

plt.show()
```

```python
# To create a quick graph using the object-oriented interface method, use the following two lines:
# Create the plot with ax.plt() 
ax = plt.axes()
ax.plot(x_axis, y_axis)

plt.show()
```

```python
# Use plt.show()  ****IMPORTANT: SHOULD ONLY BE USED ONCE PER SESSION B/C IT CAN CAUSE THE GRAPH TO DISPLAY IMPROPERLY****
#To change figure-level attributes (such as axis labels, a title, or a legend) or save the figure as an image, use fig = plt.figure().
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# 5.1.4 Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()

plt.show()
```

```python
#To declare the parameters for line color, width, and marker,
#we use color=, width=, and marker=, respectively, inside the plt.plot() function.

# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
plt.xlim(0,5)
plt.figure(figsize=(20,10))

plt.show()
```

```python
# 5.1.4 Skill Drill
#Using the ***object-oriented approach****, make the following changes to your line chart:
    #https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
fig, ax = plt.subplots()      
ax = fig.add_subplot()
#Change the line color to green."g" or "green"
#Change the marker to a diamond. "D"
#Change the line width to 2 points.
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')

ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

#Add a legend for the city of Boston.
ax.legend()
#Add a title and axes labels.
ax.set_title("PtBer Fare By Month")
#Add a y-axis limit.
ax.set_ylim(0, 45)
#Add grid lines.
ax.grid()

plt.show()
```

```python
# Create a Bubble Chart Using the MATLAB Method
    # To change the SIZE of our marker on the graph, we have to use the s, or size parameter, for the marker.
        # The s parameter must be equal to a feature of our data
                #We will make the size of each marker equal to the data floating-point decimal values of the y-axis by using s=y_axis.
plt.scatter(x_axis, y_axis)
plt.scatter(x_axis, y_axis, s=y_axis)

plt.show()
```

```python
# Increase Marker Size Parmeter Methods(2)
# 1) For loop method for increasing marker size
    # size of marker increases using for loop over y-axis data 
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)

plt.show()
```
